```python
stop_frisks = {}

for year in range(2003, 2009):
    print(year)
    df = pd.read_csv(f'../data/stop_frisk/{year}.csv', 
                                  encoding='cp437', 
                                  na_values=[' ','12311900'],
                                  dtype={'repcmd' : str, 
                                         'revcmd' : str, 
                                         'stname' : str, 
                                         'timestop' : str, 
                                         'city' : 'category', 
                                         'sector' : 'category',
                                         'post' : 'category',      # ignorable column
                                         'dettypcm' : 'category',  # ignorable column
                                         'officrid' : 'category',
                                        })
    # fix 2006 column names
    df = df.rename(columns={'adrnum' : 'addrnum',  
                            'adrpct': 'addrpct', 
                            'dettyp_c' : 'dettypcm', 
                            'rescod' : 'rescode',
                            'premtyp' : 'premtype',
                            'prenam' : 'premname',
                            'strintr' : 'stinter',
                            'strname' : 'stname',
                            'details_' : 'detailcm'})
    df = df.drop(columns=['detail1_','linecm','post','dettypecm'], errors='ignore')
    stop_frisks[year] = df # sort the columns alphabetically so we can check if the years match 
```
